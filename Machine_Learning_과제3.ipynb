## 3.1 Name Feature

```python
for dataset in train_and_test: 
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+).') 

train.head()
```

```python
# Title을 가진 사람이 몇 명이 존재하는지 성별과 함께 표현
pd.crosstab(train['Title'], train['Sex'])
# => train['title'] 이 index 부분, train['Sex']가 column 부부이다.
```

```python
for dataset in train_and_test:
    dataset['Title'] = dataset['Title'].replace(['Capt', 'Col', 'Countess', 'Don', 'Dona', 'Dr', 'Jonkheer', 'Lady', 'Major', 'Rev', 'Sir'], 'Other') 
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss') 
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
```

## 4-2. Sex Feature
male과 female로 나뉘어져 있는 것을 String data로 변형시킨다.

```python
for dataset in train_and_test:
    dataset['Sex'] = dataset['Sex'].astype(str)

    # astype() => 데이터 타입을 변경시켜준다.
```

## 4-3. Embarked Feature
train 데이터에서 NaN값이 존재

```python
train.Embarked.value_counts(dropna = False)
# NaN 값이 2개 존재한다는 것을 알 수 있다.
```

```python
for dataset in train_and_test:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
    # 여기서 fillna() 함수를 사용하여 2개의 NaN 값을 S로 채워준다.
    dataset['Embarked'] = dataset['Embarked'].astype(str)
    # astype() => 데이터의 타입 변경시켜준다.
```

## 3-4. Age Feature

NaN 값에 평균을 넣어주어 NaN 값을 없앤다.
Binning -> 여러 종류의 데이터에 대해 범위를 지정해주거나 카테고리를 통해 이전보다 작은 수의 그룹으로 만드는 기법
--> 단일성 분포의 왜곡을 막을 수 있지만, 이산화를 통한 데이터의 손실라는 단점도 존재

```python
for dataset in train_and_test:
    dataset['Age'].fillna(dataset['Age'].mean(), inplace = True)
    # inplace()는 기본 옵션이 False이다. False 옵션은 클립보드에만 저장하고 실행을 하지 않는다.
    # 옵션을 True로 설정하면 실행되어 앞의 코드가 반영된다.
    dataset['Age'] = dataset['Age'].astype(int)
    # adataset['Age']의 타입을 int로 변경시킨다.
    train['AgeBand'] = pd.cut(train['Age'], 5)
    # pd.cut()으로 동일 길이로 나누어서 범주를 만든 후 GroupBy()로 그룹별 통계량을 표시
    # train에 AgeBand 항목을 pd.cut을 사용하여 5개의 범주로 만든다.
print(train[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index = False).mean())
# as_index() 인덱스를 만들어준다. 기본 옵션 값 : False -> True 옵션은 인덱스가 print 되지 않는다.
```

```python
# Age에 범위를 나누어 수치형 데이터로 나타낸다.x
for dataset in train_and_test:
    dataset.loc[dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[dataset['Age'] > 64,'Age'] = 4
    dataset['Age'] = dataset['Age'].map({0 : 'Child', 1 : 'Young', 2 : 'Middle', 3 : 'Prime', 4 : 'Old'}).astype(str)
```

## 3-8. One-hot-encoding for categorical variables
카테고리 특징에 대해 one-hot-encoding과 train data와 lable을 분리시키는 작업을 하면 예측 모델에 학습시킬 준비 끝난다.

기계가 이해할 수 있도록 모든 데이터를 수치로 변환해주는 전처리 작업이 필수적이다.

pd.get_dummies 처리 => 결측값을 제외하고 0과 1로 구성된 더미값이 만들어진다.
결측값 처리(dummy_na = True 옵션) : Nan을 생성하여 결측값도 인코딩하여 처리해준다.

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)

train_label = train['Survived']

train_data = train.drop('Survived', axis=1)

test_data = test.drop("PassengerId", axis=1).copy()

```

# 4. 모델 설계 및 학습

예측 모델 5가지
1. Logicstic Regression
2. Support Vector Machine (SVM)
3. k-Nearest Neighbor (kNN)
4. Random Forest
5. Naive Bayes

```python
# sklearn 라이브러리 사용하여 모델 import
from sklearn.linear_model import LogisticRegression

from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.utils import shuffle
```

```python
# data 정렬되어있어 학습에 방해가 될 수도 있으므로 섞어줌
train_data, train_label = shuffle(train_data, train_label, random_state = 5)
```

```python
# 모델 학습과 평가에 대한 pipeline 생성

def train_and_test(model):
    model.fit(train_data, train_label) 
    prediction = model.predict(test_data)
    accuracy = round(model.score(train_data, train_label) * 100, 2) 
    print("Accuracy : ", accuracy, "%") 
    return prediction
```

```python
# column 추가 => 26개 => 35개
# train_and_test => column 단일값으로 추가
train["x27"] = 0
# => 27부터 35까지 추가 => 총 9개 column 추가
list_add = ['x27', 'x28', 'x29', 'x30', 'x31','x32','x33','x34','x35']
for i in list_add:
    test_data[i] = 0
    
print(test_data)
```

여기서 Accuracy 값이 가장 높은 RandomForestClassifier 모델을 사용한 fr_pred를 선정한다.

```python
# Logistic Regression
log_pred = train_and_test(LogisticRegression())
# SVM
svm_pred = train_and_test(SVC())
#kNN
knn_pred_4 = train_and_test(KNeighborsClassifier(n_neighbors = 4))
# Random Forest
rf_pred = train_and_test(RandomForestClassifier(n_estimators=100))
# Navie Bayes
nb_pred = train_and_test(GaussianNB())
```
